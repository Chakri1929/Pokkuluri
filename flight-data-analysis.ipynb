# 4. Feature Engineering

```python
import pandas as pd

#Calculating the annual average fare for every year
annual_avg_fare = df_cleaned.groupby('Year')['fare'].mean().reset_index()

# Renaming the column to show this is the average fare for each year
annual_avg_fare.rename(columns={'fare': 'annual_avg_fare'}, inplace=True)

#Calculating the year-over-year fare inflation rate
#fare inflation rate is calculated as 
annual_avg_fare['fare_inflation_rate'] = annual_avg_fare['annual_avg_fare'].pct_change() * 100

#Displaying the results
print(annual_avg_fare[['Year', 'annual_avg_fare', 'fare_inflation_rate']])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.model_selection import cross_val_score
```

```python
# Handling missing value
df = df.dropna()
```

```python
# Creating new features
df['fare_per_mile'] = df['fare'] / df['nsmiles']
df['total_passengers_route'] = df.groupby('tbl1apk')['passengers'].transform('sum')

# Encoding categorical variables
df_encoded = pd.get_dummies(df, columns=['carrier_lg', 'carrier_low'])

df[['fare_per_mile', 'total_passengers_route']].head()
```

```python
# Features and target variable
X = df[['nsmiles', 'passengers', 'large_ms', 'lf_ms', 'total_passengers_route']]
y = df['fare']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mse, r2
```

```python
# Cross-validation
cv_scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
```

```python
# Mean and standard deviation of cross-validation scores
cv_mean = cv_scores.mean()
cv_std = cv_scores.std()
```
